## Data Preprocessing 

#### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

Y = le.fit_transform(machines_df['Downtime'])
```

#### Column Transformer

```python
from sklearn.preprocessing import RobustScaler, MinMaxScaler
from sklearn.preprocessing import PowerTransformer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

robust_scaler_pipe = Pipeline(steps=[('imputer', SimpleImputer()), ('rob_scaler', RobustScaler())])
min_max_scaler_pipe = Pipeline(steps=[('imputer', SimpleImputer()), ('minmax_scaler', MinMaxScaler())])
power_transformer_pipe = Pipeline(steps=[('imputer', SimpleImputer()), ('pow_transformer', PowerTransformer())])


ct = ColumnTransformer(transformers=[('Rob_Scaler', robust_scaler_pipe, ['Hydraulic_Pressure(bar)','Coolant_Pressure(bar)','Coolant_Temperature',
                                                                     'Spindle_Speed(RPM)','Torque(Nm)']), 
                                     ('Min_Max', min_max_scaler_pipe, ['Cutting(kN)']),
                                     ('pow_transformer', power_transformer_pipe, ['Air_System_Pressure(bar)','Hydraulic_Oil_Temperature(?C)',
                                                                              'Spindle_Bearing_Temperature(?C)','Spindle_Vibration(?m)',
                                                                              'Tool_Vibration(?m)','Voltage(volts)']),
                                     ('OneHot',OneHotEncoder(drop='first',sparse_output=False,handle_unknown='ignore'),['Machine_ID','Assembly_Line_No'])],
                       remainder='passthrough', sparse_threshold=0)
ct
```

## Data Splitting

```python
X = machines_df.drop(columns='Downtime')

X.shape, Y.shape
```

## Model Selection

```python
from sklearn.mixture import GaussianMixture, BayesianGaussianMixture
from sklearn.cluster import KMeans, BisectingKMeans, MiniBatchKMeans, DBSCAN, Birch, SpectralClustering, AgglomerativeClustering, AffinityPropagation

models = {
    'gauss_mix': GaussianMixture(n_components=2),
    'bayes_gauss_mix': BayesianGaussianMixture(n_components=2),
    'kmeans': KMeans(n_clusters=2),
    'bisect_kmeans': BisectingKMeans(n_clusters=2),
    'min_batch_kmeans': MiniBatchKMeans(n_clusters=2),
    'dbscan': DBSCAN(),
    'birch': Birch(n_clusters=2),
    'spect_clust': SpectralClustering(n_clusters=2),
    'agglomerative': AgglomerativeClustering(n_clusters=2),
    'affinity': AffinityPropagation()
}
```

#### Cross Validations

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer, rand_score

X_pr = ct.fit_transform(X)
scores_df = pd.DataFrame(index=models.keys(), columns=['Rand_Scores', 'Mean_Rand_Score'])

for i in models.keys():
    scores = cross_val_score(models[i], X_pr, Y, cv = 5, scoring=make_scorer(rand_score))
    scores_df.loc[i, 'Rand_Scores'] = scores
    scores_df.loc[i, 'Mean_Rand_Score'] = scores.mean()
    print(f'{i} --> cross validated')
```

```python
scores_df.sort_values(by = 'Mean_Rand_Score', ascending=False)
```

```python
demo_model = DBSCAN()
y_pred = demo_model.fit_predict(X_pr)
print('Score: ', rand_score(Y, y_pred))
```

```python
demo_model = AgglomerativeClustering(n_clusters=2)
y_pred = demo_model.fit_predict(X_pr)
print('Score: ', rand_score(Y, y_pred))
```

```python
demo_model = SpectralClustering(n_clusters=2)
y_pred = demo_model.fit_predict(X_pr)
print('Score: ', rand_score(Y, y_pred))
```

## Model Building

```python
model = Pipeline(steps=[('transformer', ct), ('agglo_clustering', AgglomerativeClustering(n_clusters=2))])

model.fit(X)
```

## Model Evaluation

```python
from sklearn.metrics import rand_score, normalized_mutual_info_score, silhouette_score, davies_bouldin_score, calinski_harabasz_score

y_pred = model.fit_predict(X)
y_labels = model.steps[1][1].labels_
evaluation_df = pd.DataFrame(index=['rand_score','normalized_mutual_info_score','silhouette_score','davies_bouldin_score','calinski_harabasz_score'], 
                             columns=['Score'])

evaluation_df.loc['rand_score','Score'] = rand_score(Y, y_pred)
evaluation_df.loc['normalized_mutual_info_score','Score'] = normalized_mutual_info_score(Y, y_pred)
evaluation_df.loc['silhouette_score','Score'] = silhouette_score(X_pr, y_labels)
evaluation_df.loc['davies_bouldin_score','Score'] = davies_bouldin_score(X_pr, y_labels)
evaluation_df.loc['calinski_harabasz_score','Score'] = calinski_harabasz_score(X_pr, y_labels)
```
